```python
import csv
import pandas as pd
import os
import cv2

images = "images/"
digits = ['#', '$', '%', '&', '1', '2', '3', '4', '5']
```

```python

import numpy as np

f = open('data.csv', 'w', encoding='utf8', newline='')

writer = csv.writer(f)

headers = ['digito', 'densidad', 'alto', 'ancho']

writer.writerow(headers)

for image in os.listdir(images):
    path = f"{images}{image}"
    img = cv2.imread(path, 0)

    digito = image.split(' ')[0]

    densidad = 0
    x, y = img.shape[:2]

    alto_max = x
    alto_min = 0
    ancho_max = y
    ancho_min = 0

    for i in range(x):
        for j in range(y):
            densidad += img[i, j]

            if img[i, j] < 255 // 2:
                alto_max = i if alto_max > i else alto_max
                alto_min = i if i > alto_min else alto_min

                ancho_max = j if ancho_max > j else ancho_max
                ancho_min = j if j > ancho_min else ancho_min

    alto = abs(alto_max - alto_min)
    ancho = abs(ancho_max - ancho_min)

    densidad = densidad / (x * y * 255)
    alto = alto / x
    ancho = ancho / y

    row = [digito, densidad, alto, ancho]

    writer.writerow(row)

f.close()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

dataframe = pd.read_csv('data.csv', sep=',')
dataframe.head()
```

```python
from matplotlib import pyplot as plt

dataframe.hist()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X = dataframe[['densidad', 'alto', 'ancho']].values
y = dataframe['digito'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python

from sklearn.neighbors import KNeighborsClassifier

n_neighbors = 7

knn = KNeighborsClassifier(n_neighbors)
knn.fit(X_train, y_train)
print('Accuracy {:.2f}'.format(knn.score(X_test, y_test)))

```

```python

k_range = range(1, len(y) // 10)
scores = []
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    scores.append(knn.score(X_test, y_test))
plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
```
